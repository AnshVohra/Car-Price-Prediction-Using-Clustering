```python
def plot_count(x,fig):
    plt.subplot(4,2,fig)
    plt.title(x+' Histogram')
    sns.countplot(df[x],palette=("magma"))
    plt.subplot(4,2,(fig+1))
    plt.title(x+' vs Price')
    sns.boxplot(x=df[x], y=df.Price_Ex, palette=("magma"))
    
plt.figure(figsize=(15,20))

plot_count('Engine_Location', 1)
plot_count('Cylinders', 3)
plot_count('Fuel_System', 5)
plot_count('Drivetrain', 7)

plt.tight_layout()
```
